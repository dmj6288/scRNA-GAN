```python

import pandas as pd
import numpy as np
import itertools
import pickle 

# File path
gct_file = "GTEx_Analysis_2017-06-05_v8_RSEMv1.3.0_transcript_tpm.gct"

# Define chunk size
chunk_size = 10000
```

```python
brain_image_files = pd.read_csv("../ref_files/gtex_stomach.csv")
protein_coding_genes = pd.read_csv("../ref_files/protein_coding_genes.csv")
brain_image_samples = list(brain_image_files['wsi_file_name'])[1:]
```

```python

with open("gtex_columns.pkl", "rb") as file:
    sample_names = pickle.load(file)
    
```

```python
cortex_indices_list_of_lists = [[i+2 for i in range(len(sample_names)) if sample_names[i]==brain_image_sample] for brain_image_sample in brain_image_samples]
all_liver_sample_indices     = [0, 1] + list(itertools.chain.from_iterable(cortex_indices_list_of_lists))
```

```python

with open("gtex_column_names_stomach.pkl", "wb") as file:
    pickle.dump(all_liver_sample_indices, file)
    
```

```python

chunk_size = 1000
```

```python

# Skip the first two metadata lines
with pd.read_csv(gct_file, sep="\t", skiprows=2, chunksize=chunk_size) as reader:
           
    for i, chunk in enumerate(reader):
        
        chunk['gene_id'] = 'rna_' + chunk['gene_id'].astype(str)
        chunk            = chunk[chunk['gene_id'].isin(list(protein_coding_genes['Gene']))]
        
        if i == 0:
            current_chunk = chunk.iloc[:, all_liver_sample_indices]
        
        if i > 0 and chunk[chunk['gene_id'].isin(list(protein_coding_genes['Gene']))].shape[0] > 0:
            current_chunk = current_chunk.append(chunk.iloc[:, all_liver_sample_indices])
            
        print(i)
        print(chunk.shape)
        print(current_chunk.shape)
        print(f"Processing chunk {i+1}")
        
final_aggregated_liver_samples = current_chunk.groupby('gene_id').sum()

with open("stomach_samples.pkl", "wb") as file:
    pickle.dump(final_aggregated_liver_samples, file)
```

```python

with open("stomach_samples.pkl", "rb") as file:
    data = pickle.load(file)
```
