```python
import pandas as pd
import regex as re
```

```python
gtex_lung = pd.read_csv("../ref_files/gtex_lung.csv")
```

```python
lung_image_samples = list(gtex_lung['wsi_file_name'])[1:]
```

```python

with open("../../Images/lung/lung_image_urls.txt", "w") as file:
    for lung_image_sample in lung_image_samples:
        #print(brain_image_sample)

        # Split the string into parts
        parts = re.split(r'[-.\s]+', lung_image_sample)

        # Increment the numeric part
        parts[2] = str(int(parts[2]) + 1).zfill(len(parts[2]))

        # Rejoin the parts into a single string
        source_link = "https://brd.nci.nih.gov/brd/imagedownload/" + '-'.join(parts[:3])
        file.write(source_link + "\n")
        print(source_link)
        #updated_brain_image_samples.append(new_string)
```
