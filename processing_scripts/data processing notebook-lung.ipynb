```python

import pandas as pd
import numpy as np
import itertools
import pickle 
import re

# File path
gct_file = "GTEx_Analysis_2017-06-05_v8_RSEMv1.3.0_transcript_tpm.gct"

```

```python
lung_image_files = pd.read_csv("../ref_files/gtex_lung.csv")
protein_coding_genes = pd.read_csv("../ref_files/protein_coding_genes.csv")
lung_image_samples = list(lung_image_files['wsi_file_name'])[1:]
```

```python
len(lung_image_samples)
```

```python

with open("processing_intermediate_files/gtex_columns.pkl", "rb") as file:
    sample_names = pickle.load(file)
    
```

```python
updated_lung_image_samples = []
for lung_image_sample in lung_image_samples:
    #print(brain_image_sample)
    
    # Split the string into parts
    parts = re.split(r'[-.\s]+', lung_image_sample)

    # Increment the numeric part
    parts[2] = str(int(parts[2]) + 1).zfill(len(parts[2]))

    # Rejoin the parts into a single string
    new_string = '-'.join(parts[:3])
    updated_lung_image_samples.append(new_string)
```

```python
updated_lung_sample_names = ["-".join(sample_name.split("-")[:3]) for sample_name in sample_names]
```

```python
lung_indices_list_of_lists = [[i+2 for i in range(len(updated_lung_sample_names)) if updated_lung_sample_names[i]==lung_image_sample] for lung_image_sample in updated_lung_image_samples]
all_lung_sample_indices     = [0, 1] + list(itertools.chain.from_iterable(lung_indices_list_of_lists))
```

```python
len(all_lung_sample_indices)
```

```python

with open("processing_intermediate_files/gtex_column_names_lung.pkl", "wb") as file:
    pickle.dump(all_lung_sample_indices, file)
    
```

```python

chunk_size = 1000
```

```python

# Skip the first two metadata lines
with pd.read_csv(gct_file, sep="\t", skiprows=2, chunksize=chunk_size) as reader:
           
    for i, chunk in enumerate(reader):
        
        chunk['gene_id'] = 'rna_' + chunk['gene_id'].astype(str)
        chunk            = chunk[chunk['gene_id'].isin(list(protein_coding_genes['Gene']))]
        
        if i == 0:
            current_chunk = chunk.iloc[:, all_lung_sample_indices]
        
        if i > 0 and chunk[chunk['gene_id'].isin(list(protein_coding_genes['Gene']))].shape[0] > 0:
            current_chunk = current_chunk.append(chunk.iloc[:, all_lung_sample_indices])
            
        print(i)
        print(chunk.shape)
        print(current_chunk.shape)
        print(f"Processing chunk {i+1}")

        
```

```python
final_aggregated_lung_samples = current_chunk.groupby('gene_id').sum()

with open("processing_intermediate_files/lung_samples.pkl", "wb") as file:
    pickle.dump(final_aggregated_lung_samples, file)
```

```python

final_aggregated_lung_samples.to_csv("data_for_beta_VAE/GTex_Lung_data_SSL_proteincoding.csv")
```
